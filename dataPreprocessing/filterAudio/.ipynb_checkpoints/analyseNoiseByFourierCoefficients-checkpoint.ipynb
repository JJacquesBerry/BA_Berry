```python
import librosa.display
import numpy as np
```

Load Audiodata

```python
audioPath = "../soundFiles/TroglodytesTroglodytes/WAV/TroglodytesTroglodytes0.wav"
#audioPath = "soundFiles/Test/test1.m4a"
    
audioData, sr = librosa.load(audioPath)
     
print(sr)
print(len(audioData))
```

Get noise
- shorten noise to one second

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(20, 3))
plt.plot(audioData)

plt.figure(figsize=(20, 3))
plt.plot(audioData[44100:88200])

noiseArray = np.array([])
noiseArray = np.append(noiseArray, audioData[0:11500])
noiseArray = np.append(noiseArray, audioData[22000:38500])
noiseArray = np.append(noiseArray, audioData[49000:59500])
noiseArray = np.append(noiseArray, audioData[68000:80000])

oneSecondNoise = noiseArray[0:22050]
print(len(oneSecondNoise))

plt.figure(figsize=(20, 3))
plt.plot(oneSecondNoise)
```

Fourier-transform one second of noise

```python
noiseStft = librosa.stft(oneSecondNoise)

print(noiseStft.shape)
print(noiseStft[0])
```

Visualize the noise

```python
plt.figure(figsize=(20, 10))
for plot in noiseStft:
    plt.plot(abs(plot))

plt.figure(figsize=(20, 3))
plt.plot(oneSecondNoise)
```

1. Filter the noise by fourier-coefficients
- search and nullify relevant fourier-coefficients

```python
import soundfile as sf

sf.write("../soundFiles/readSoundfiles/filterNoise/filterByFourierCoefficients/oneSecondNoise.wav", oneSecondNoise, sr)
```

2. Filter low values
- save as a file

```python
print(len(noiseStft))
```

```python
def filterFC(fcArray):
    fcArrayCopy = fcArray.copy()

    for i in range(len(fcArray)):
        subArray = fcArray[i]
        for j in range(len(subArray)):
            if abs(subArray[j]) <= 0.3:
                fcArrayCopy[i][j] = 0+0.j

    return(fcArrayCopy)
```

```python
filteredFC = filterFC(noiseStft)

plt.figure(figsize=(20, 10))
for plot in filteredFC:
    plt.plot(abs(plot))
```

Get sound from filtered FC

```python
filteredFCSound = librosa.istft(filteredFC)

print(filteredFCSound)

sf.write("../soundFiles/readSoundfiles/filterNoise/filterByFourierCoefficients/oneSecondNoiseFiltered-0_3.wav", filteredFCSound, sr)
```
