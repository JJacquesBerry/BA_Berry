Load Audiodata

```python
audioPath = "../../soundFiles/TroglodytesTroglodytes/WAV/TroglodytesTroglodytes0.wav"
#audioPath = "../soundFiles/Test/test1.m4a"
    
audioData, sr = librosa.load(audioPath, sr=11025, duration=5)
    
print(sr)
print(len(audioData))
```

STFT (Short-time fourier transform)

```python
#https://librosa.org/doc/latest/generated/librosa.stft.html
stftData = librosa.stft(audioData)
print(stftData.shape)
```

Examining Data

2. stft-Data

2.1 length

The dimension is always [1025, *], where * is:


Sr-Variation

bei sr 1000 und length 1 ist width = 2

bei sr 2000 und length 1 ist width = 4

bei sr 3000 und length 1 ist width = 6

formula: stftWidth = sr/500


Duration-variation

bei sr 1000 und length 0.5 ist width = 1

bei sr 1000 und length 1 ist width = 2

bei sr 1000 und length 1.5 ist width = 3

bei sr 1000 und length 2 ist width = 4

formula: stftWidth = length/500


Gesamt-Variation

formula: stftWidth = (sr * length)/500
