```python
import librosa.display
import matplotlib.pyplot as plt
import numpy as np
```

```python
#get audio

audioPath = "../soundFiles/TroglodytesTroglodytes/WAV/TroglodytesTroglodytes0.wav"
#audioPath = "../soundFiles/Test/test1.m4a"
audioData, sr = librosa.load(audioPath)
```

Stft

```python
    #https://librosa.org/doc/latest/generated/librosa.stft.html
stftData = librosa.stft(audioData)
print(stftData[0][0])
```

Stftf-Data Options

abs
- returns an absolute value of a number

```python
    #https://numpy.org/doc/stable/reference/generated/numpy.absolute.html
print(np.abs(stftData[0][0]))
```

angle
- returns an angle

```python
    #https://numpy.org/doc/stable/reference/generated/numpy.angle.html
print(np.angle(stftData[0][0]))
```

Amplitude

```python
    #https://librosa.org/doc/latest/generated/librosa.amplitude_to_db.html
stftDataAmplitude = librosa.amplitude_to_db(abs(stftData))
```

```python
print(len(stftDataAmplitude))
print(len(stftDataAmplitude[0]))
```

Draw the spectrogram

```python
#with information
plt.figure(figsize=(10, 5))
librosa.display.specshow(stftDataAmplitude, sr=sr, x_axis='time', y_axis='hz')
plt.colorbar()
```

```python
#without information
plt.figure(figsize=(10, 9))
librosa.display.specshow(stftDataAmplitude, sr=sr)
```
